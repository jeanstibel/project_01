```python
def preprocess_data(data):
	# reset index to keep crypto_id
	data.reset_index(inplace=True)

	# group by date and get the mean of the other columns
	data = data.groupby(['date', 'crypto_id']).agg({
		'open': 'first',                        # get the first value of the day
		'high': 'max',                          # get the amount higher value of the day
		'low': 'min',                           # get the amount lower value of the day
		'close': 'last',                        # get the last value of the day
		'volume': 'sum',                        # get the sum of the volume of the day
		'quote_asset_volume': 'sum',            # get the sum of the quote_asset_volume of the day
		'number_of_trades': 'sum',              # get the sum of the number_of_trades of the day
		'taker_buy_base_asset_volume': 'sum',   # get the sum of the taker_buy_base_asset_volume of the day 
		'taker_buy_quote_asset_volume': 'sum'   # get the sum of the taker_buy_quote_asset_volume of the day
	}).reset_index()

	# Reorganize the columns
	data = data[['crypto_id', 'date', 'open', 'high', 
				 'low', 'close', 'volume', 'quote_asset_volume', 
				 'number_of_trades', 'taker_buy_base_asset_volume',
				 'taker_buy_quote_asset_volume']]
	
	return data


data.head(20)
```
